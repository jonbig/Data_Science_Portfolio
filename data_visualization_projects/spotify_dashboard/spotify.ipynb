```python
import pandas as pd
import json
import numpy as np
```

```python
# list of json files
json_files = ['endsong_0.json', 'endsong_1.json', 'endsong_2.json', 'endsong_3.json', 'endsong_4.json', 'endsong_5.json', 'endsong_6.json', 'endsong_7.json', 'endsong_9.json', 'endsong_9.json']
```

```python
# initialize an empty dataframe
df = pd.DataFrame()
```

```python
# checking the encoding of the json files
import chardet

def detect_file_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

file_path = 'endsong_0.json'
encoding = detect_file_encoding(file_path)

print(f'The detected encoding of {file_path} is {encoding}')
```

```python
#iterate over the json files, combine ihto pandas df

for file in json_files:
    with open(file, 'r', encoding='utf8') as f:
        # load json data
        data = json.load(f)
        
        # convert json data into a dataframe
        temp_df = pd.json_normalize(data)  # use pd.json_normalize for nested json
        
        # append the data to the main dataframe
        df = df.append(temp_df, ignore_index=True)

print(df)
```
